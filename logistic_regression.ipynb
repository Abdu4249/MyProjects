```python
from sklearn.model_selection import train_test_split
from sklearn.metrics import confusion_matrix,accuracy_score,classification_report
from sklearn.linear_model import LogisticRegression
```

```python
# Splitting Dataset to Train and Test data
X=df_final.drop("Label",axis=1)
Y=df_final['Label']
xtrain,xtest,ytrain,ytest=train_test_split(X,Y,test_size=0.2,random_state=42)
```

```python
# Convert feature names to strings
xtrain.columns = xtrain.columns.astype(str)
xtest.columns = xtest.columns.astype(str)
```

```python
# Modeling using Logistic Regression
lr=LogisticRegression(max_iter=100000)
lr.fit(xtrain,ytrain)
ypredicted=lr.predict(xtest)
```

```python
#printing confusion matrix
print("Confusion Matrix:")
print("  ")
print(confusion_matrix(ytest,ypredicted))
print("---------------")
print("  ")
print("Accuracy Score:")
print("  ")
print(accuracy_score(ytest,ypredicted))
print("  ")
print("---------------")
print("  ")
print("Classification Report:")
print("  ")
print(classification_report(ytest,ypredicted))
```

```python
# It is clear from above all modelling that there is a target class imbalance
# To address class imbalance we use SMOTE Method
from imblearn.combine import SMOTETomek
from collections import Counter
sm=SMOTETomek()
# Convert feature names in X to strings before resampling
X.columns = X.columns.astype(str) # Change feature names to strings for the original DataFrame X
X_sm,Y_sm=sm.fit_resample(X,Y)
print("The number of classes before resampling {}".format(Counter(Y)));
print("The number of classes after resampling {}".format(Counter(Y_sm)));
```

```python
# Splitting Dataset to Train and Test data
xtrainsm,xtestsm,ytrainsm,ytestsm=train_test_split(X_sm,Y_sm,test_size=0.25,random_state=42,stratify=Y_sm)
```

```python
# Logistic regression after over sampling
lr.fit(xtrainsm,ytrainsm)
ypredictedsm=lr.predict(xtestsm)

#printing confusion matrix
print("Confusion Matrix:")
print("  ")
print(confusion_matrix(ytestsm,ypredictedsm))
print("---------------")
print("  ")
print("Accuracy Score:")
print("  ")
print(accuracy_score(ytestsm,ypredictedsm))
print("  ")
print("---------------")
print("  ")
print("Classification Report:")
print("  ")
print(classification_report(ytestsm,ypredictedsm))
```

```python
# scaled and OVER SAMPLED
from sklearn import preprocessing
min_max_scaler = preprocessing.MinMaxScaler()
x_scaled=min_max_scaler.fit_transform(X)
```

```python
X_smsc,Y_smsc=sm.fit_resample(x_scaled,Y)
print("The number of classes before resampling {}".format(Counter(Y)))
print("The number of classes after resampling {}".format(Counter(Y_smsc)))
```

```python
xtrainsmsc,xtestsmsc,ytrainsmsc,ytestsmsc=train_test_split(X_smsc,Y_smsc,test_size=0.25,random_state=42,stratify=Y_smsc)
```
